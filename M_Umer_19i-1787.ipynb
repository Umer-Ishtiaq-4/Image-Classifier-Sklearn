```python
from skimage.io import imread, imshow
from skimage.transform import resize
from skimage.feature import hog
from skimage import exposure
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Load Data

```python
Xtr=np.loadtxt("TrainData.csv")
Ytr=np.loadtxt("TrainLabels.csv")
Xts=np.loadtxt("TestData.csv")
```

```python
#reading the image
img = imread('car.jpg')
imshow(img)
print(img.shape)
## Extracting Hog Features
fd, hog_image = hog(img, orientations=9, pixels_per_cell=(8, 8), 
                    cells_per_block=(2, 2), visualize=True, multichannel=True)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6), sharex=True, sharey=True) 

ax1.imshow(img, cmap=plt.cm.gray) 
ax1.set_title('Input image') 
ax1.set_xlabel('M Umer 19I-1787')

# Rescale histogram for better display 
hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range=(0, 10)) 
ax2.imshow(hog_image_rescaled, cmap=plt.cm.gray) 
ax2.set_title('Histogram of Oriented Gradients')

plt.show()
```

## Step 4

```python
#plt.imshow(Xtr[10].reshape([28,28]))
shirt =Xtr[10].reshape([28,28])
## Extracting Hog Features
fd, shirt_img = hog(shirt, orientations=8, pixels_per_cell=(4, 4),cells_per_block=(1, 1), visualize=True, multichannel=None)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8,4), sharex=True, sharey=True) 

ax1.axis('off')
ax1.imshow(shirt, cmap=plt.cm.gray) 
ax1.set_title('Input image') 

# Rescale histogram for better display 
hog_image_rescaled = exposure.rescale_intensity(shirt_img, in_range=(0, 10))
ax2.axis('off')
ax2.imshow(hog_image_rescaled, cmap=plt.cm.gray) 
ax2.set_title('Histogram of Oriented Gradients')
plt.show()
```

## Step 5

```python
def Extract(shirt):
    fd = hog(shirt, orientations=8, pixels_per_cell=(4, 4),cells_per_block=(1, 1), visualize=False, multichannel=None)
    return fd
```

```python
x_hog = []
y_hog = []
for i in range(len(Xtr)):
    #X_feature = Xtr[i]
    X_feature = Extract(Xtr[i].reshape([28,28]))
    x_hog.append(X_feature)
    y_hog.append(Ytr[i])
```

```python
#print(len(x_hog[0]))
```

## Step 6 Splitting

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_hog, y_hog, test_size = 0.20, random_state = 0)
```

```python
print(len(X_train))
print(len(X_test))
```

## Cross Validation Score

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
```

## 1) SVM

```python
from sklearn.svm import SVC
svc = SVC(kernel='linear',gamma=0.001,C=1)
```

```python
cross_val_score(svc,X_train,y_train,cv=5)
```

## 2) Deicssion Tree

```python
from sklearn.tree import DecisionTreeClassifier 
clf = DecisionTreeClassifier(criterion="entropy", max_depth=3)
```

```python
cross_val_score(clf,X_train,y_train,cv=5)
```

## 3) Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier(max_depth=2, random_state=0)
```

```python
cross_val_score(RF,X_train,y_train,cv=5)
```

## Step 7

```python
y_pred_dt = clf.predict(X_test)
```

## Decission Tree Accuracy
